# Data Cleaning

## 1) Convert Year column to integer

```python
df['Year']=df['Year'].astype(int)
```

### Null values must be removed first

```python
df=df.dropna(subset=['Year'])
```

```python
df['Year'].isna().sum()
```

```python
df['Year']=df['Year'].astype(int)
```

```python
df['Year'].dtype
```

## 2) Convert Group column to integer

```python
df['Group'].isna().sum()
```

```python
df['Group']=df['Group'].astype(int)
```

```python
df['Group'].dtype
```

## 3) Convert Birth Date column to datetime

```python
df['Birth Date'].isna().sum()
```

```python
df['Birth Date']=pd.to_datetime(df['Birth Date'])
```

```python
df['Birth Date'].sample()
```

## 4) Convert Death Date column to datetime

```python
df['Death Date'].isna().sum()
```

```python
df.shape
```

### The majority of the Death Date column is null therefore, they should be handled in a way other than dropping the rows because that would mean dropping most of the dataframe

```python
df['Death Date'] = pd.to_datetime(df['Death Date'], errors='coerce')
```

```python
df['Death Date'].sample()
```

# Which year has the most Astronauts hired?

```python
years=df['Year'].value_counts().sort_index().plot(kind='barh', color=['#e345c6', '#1bd337'])
years.set_xlabel('Number of Astronauts hired')
years.set_ylabel('Year');
```

## Insight: 1996 and 1978 had the most Astronauts hired (35 Astronauts)

# What is the average age of Astronauts when they get hired?

```python
df['Birth Year']=df['Birth Date'].dt.year     #Create column that has year only from Birth Date
```

```python
df.sample()
```

```python
df['Age']=df['Year']-df['Birth Year']     #Create column that subtracts the year the Astronaut was born from the year the Astronaut got hired to get Age
```

```python
df.sample()
```

```python
ages=df['Age'].value_counts().sort_index().plot(kind='bar', color=['#29d5d3', '#d59e29'])
ages.set_xlabel('Age')
ages.set_ylabel('Number of Astronauts');
```

```python
df['Age'].mean()
```

## Insight: Average age of Astronauts when they get hired is 34

# Where did Astronauts study? 

```python
uni=df[['Alma Mater']]
```

```python
uni
```

```python
uni=uni['Alma Mater'].str.split(';', expand=True)
uni                                              #Separate each university name from each row
```

```python
uni0=uni[0]
uni1=uni[1]
uni2=uni[2]
uni3=uni[3]
uni4=uni[4]
uni5=uni[5]
```

```python
unis=pd.DataFrame(pd.concat([uni0, uni1, uni2, uni3, uni4, uni5]))        #Adding all university names in one dataframe
unis
```

```python
unis.rename(columns = {0 :'University'}, inplace = True)
unis
```

```python
unis['University']=unis['University'].str.strip()       #Remove spaces before and after the university name
```

```python
unis.dropna(inplace=True) 
```

```python
unis.isna().sum()
```

```python
unis
```

```python
unis_count=pd.DataFrame(unis['University'].value_counts())
unis_count
```

```python
unis_count.reset_index(inplace=True)
```

```python
unis_count
```

```python
unis_count=unis_count[unis_count['count']>10]
unis_count
```

```python
uni_plot=unis_count.plot(kind='barh', x='University', y='count', color=['#aa8718', '#e5ae13'], legend=False)
uni_plot.set_xlabel('Number of Astronauts')
uni_plot.set_ylabel('University');
```
